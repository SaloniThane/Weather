```python
import pandas as pd
df = pd.read_csv('/content/aerofit_treadmill.csv')
```

```python
df.head(2)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Product'] = df['Product'].astype('category')
df['Gender'] = df['Gender'].astype('category')
df['MaritalStatus'] = df['MaritalStatus'].astype('category')
```

```python
df.info()
```

```python
print("Product Counts:\n", df['Product'].value_counts())
print("\nGender Counts:\n", df['Gender'].value_counts())
print("\nMarital Status Counts:\n", df['MaritalStatus'].value_counts())
```

```python
unique_values = {col: df[col].unique() for col in df.columns}
for col, unique_vals in unique_values.items():
    print(f"Unique values in {col}:{unique_vals}\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))

for col in ['Age', 'Usage', 'Fitness', 'Income', 'Miles']:
    plt.figure(figsize=(6, 3))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Product', data=df)
plt.title('Count of Treadmills Purchased')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df)
plt.title('Count of Customers by Gender')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='MaritalStatus', data=df)
plt.title('Count of Customers by Marital Status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product', y='Income', data=df)
plt.title('Income Distribution by Product')
plt.show()
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product', y='Miles', data=df)
plt.title('Miles Distribution by Product')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr_matrix = df[['Age', 'Usage', 'Fitness', 'Income', 'Miles']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Usage', 'Fitness', 'Income', 'Miles']])
plt.title('Pairplot of Continuous Variables')
plt.show()
```

```python
print("Missing Values in Each Column:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Age', 'Usage', 'Fitness', 'Income', 'Miles']])
plt.title("Outlier Detection Using Boxplot")
plt.show()
```
